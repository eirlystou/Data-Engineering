## 2. PDF í˜ì´ì§€ ë¶„í•  ë° Google Vision OCR ì‹¤í–‰

ì´ ë‹¨ê³„ì—ì„œëŠ” ìŠ¤ìº”ëœ PDF íŒŒì¼ì„ ì´ë¯¸ì§€ë¡œ ë³€í™˜í•˜ê³ , ê° í˜ì´ì§€ì—ì„œ í…ìŠ¤íŠ¸ë¥¼ ì¶”ì¶œ(OCR)í•©ë‹ˆë‹¤.

- ì²« ë²ˆì§¸ í˜ì´ì§€ëŠ” ì•ˆë‚´ë¬¸ì´ë¯€ë¡œ ì „ì²´ OCR ìˆ˜í–‰
- 2í˜ì´ì§€ë¶€í„°ëŠ” 2ë‹¨ ë ˆì´ì•„ì›ƒì„ ê³ ë ¤í•˜ì—¬ ì¢Œ/ìš°ë¡œ ë‚˜ëˆ„ì–´ OCR ìˆ˜í–‰
- ìƒë‹¨(Header)ê³¼ í•˜ë‹¨(Footer)ì€ ë¹„ìœ¨ ê¸°ì¤€ìœ¼ë¡œ ì œê±°í•˜ì—¬ ë…¸ì´ì¦ˆë¥¼ ì¤„ì„
- OCR ê²°ê³¼ëŠ” `output/page_ë²ˆí˜¸.txt` í˜•íƒœë¡œ ì €ì¥

ğŸ“Œ ì‚¬ìš© ê¸°ìˆ :
- `pdf2image` : PDF â†’ ì´ë¯¸ì§€ ë³€í™˜
- `OpenCV` : ì´ë¯¸ì§€ ìë¥´ê¸° ë° ì €ì¥
- `Google Cloud Vision API` : OCR (ë¬¸ì„œ í…ìŠ¤íŠ¸ ì¸ì‹)

```python
run_ocr()           # Vision APIë¥¼ ì´ìš©í•œ í…ìŠ¤íŠ¸ ì¶”ì¶œ í•¨ìˆ˜

````python
import os
import cv2
import io
from pdf2image import convert_from_path
from google.cloud import vision

# API í‚¤ ê²½ë¡œ ì„¤ì •
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/ntlkbuai-9302f377d55f.json"  # âš ï¸ íŒŒì¼ ê²½ë¡œë¥¼ ì •í™•íˆ ì§€ì •í•˜ì„¸ìš”

#ì´ë¯¸ì§€ ì €ì¥ í´ë” ìƒì„±
os.makedirs("pages", exist_ok=True)
os.makedirs("output", exist_ok=True)

# Google Vision OCR í•¨ìˆ˜ ì •ì˜
def run_ocr(image_path):
    with io.open(image_path, 'rb') as image_file:
        content = image_file.read()
    image = vision.Image(content=content)
    response = vision_client.document_text_detection(image=image)
    return response.full_text_annotation.text

# ì´ë¯¸ì§€ì—ì„œ í—¤ë”/í‘¸í„° ì œê±° í›„ ì¢Œ/ìš°ë¡œ ë¶„í• 
def crop_and_split_image(img_path, header_ratio=0.12, footer_ratio=0.1):
    img = cv2.imread(img_path)
    h, w = img.shape[:2]

    top = int(h * header_ratio)
    bottom = h - int(h * footer_ratio)
    main_body = img[top:bottom, :]

    mid = w // 2
    left = main_body[:, :mid]
    right = main_body[:, mid:]

    prefix = img_path.replace(".png", "")
    left_path = f"{prefix}_left.png"
    right_path = f"{prefix}_right.png"

    cv2.imwrite(left_path, left)
    cv2.imwrite(right_path, right)

    return left_path, right_path

# Vision API í´ë¼ì´ì–¸íŠ¸ ì´ˆê¸°í™”
vision_client = vision.ImageAnnotatorClient()

# PDF íŒŒì¼ì„ ì´ë¯¸ì§€ë¡œ ë³€í™˜
pdf_path = "/content/2020ë…„ë„ ì œ48íšŒ ë¬¼ë¦¬ì¹˜ë£Œì‚¬ êµ­ê°€ì‹œí—˜_1êµì‹œ(í™€ìˆ˜í˜•).pdf"
images = convert_from_path(pdf_path, dpi=300)

# ê° í˜ì´ì§€ë¥¼ OCR ì²˜ë¦¬
for i, img in enumerate(images):
    page_num = i + 1
    img_path = f"pages/page_{page_num}.png"
    img.save(img_path)

    if page_num == 1:
        # ì²« í˜ì´ì§€ëŠ” ë¶„í•  ì—†ì´ ì „ì²´ OCR
        full_text = run_ocr(img_path)
    else:
        # 2í˜ì´ì§€ë¶€í„°ëŠ” ì¢Œìš°ë¡œ ë‚˜ëˆ„ì–´ OCR
        left_img, right_img = crop_and_split_image(img_path)
        text_left = run_ocr(left_img)
        text_right = run_ocr(right_img)
        full_text = f"{text_left.strip()}\n{text_right.strip()}"

    # í…ìŠ¤íŠ¸ íŒŒì¼ë¡œ ì €ì¥
    with open(f"output/page_{page_num}.txt", "w", encoding="utf-8") as f:
        f.write(full_text)

    print(f"âœ… í˜ì´ì§€ {page_num} ì²˜ë¦¬ ì™„ë£Œ")
````

## ğŸ§  3. ì§ˆë¬¸ ë° ë³´ê¸°ë¥¼ ë¸”ë¡ì—ì„œ ë¶„ë¦¬

ì´ ë‹¨ê³„ì—ì„œëŠ” ê° ë¬¸í•­ ë¸”ë¡ì„ ì²˜ë¦¬í•˜ì—¬:
- ë¬¸ì œ ë³¸ë¬¸ê³¼ ë³´ê¸°(â‘ ~â‘¤ ë˜ëŠ” 1~5)ë¥¼ êµ¬ë¶„í•˜ê³ 
- ë³´ê¸° ë²ˆí˜¸ëŠ” ì œê±°í•œ í›„ ë‚´ìš©ë§Œ ë‚¨ê¹ë‹ˆë‹¤.
- ìµœì¢…ì ìœ¼ë¡œ `question_id`, `ì§ˆë¬¸`, `ë³´ê¸°ê°€ ë‹´ê¸´ ë¦¬ìŠ¤íŠ¸` í˜•ì‹ìœ¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤.

ğŸ“Œ ì£¼ìš” ì²˜ë¦¬:
- `dap_an_start`: ì •ë‹µì´ ì‹œì‘ë˜ëŠ” ì¤„ ë²ˆí˜¸ ì°¾ê¸°
- ê¸€ë¨¸ë¦¬ ê¸°í˜¸(`â€¢`, `â—‹`, `-` ë“±)ê°€ ìˆìœ¼ë©´ ì¤„ ë‹¨ìœ„ë¡œ ë‚˜ëˆ„ê³ , ì—†ìœ¼ë©´ ì´ì „ ì¤„ì— ì´ì–´ ë¶™ì„
- ì •ë‹µ ë¶€ë¶„ì€ ì •ê·œí‘œí˜„ì‹ìœ¼ë¡œ â‘ , â‘¡, â‘¢ ë“±ìœ¼ë¡œ ë¶„ë¦¬
- ë²ˆí˜¸(â‘ , 1 ë“±)ëŠ” ì œê±° í›„ ë‚´ìš©ë§Œ ì €ì¥

ğŸ“¤ ê²°ê³¼ëŠ” `all_questions` ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€ë˜ë©°, ì´í›„ JSON íŒŒì¼ë¡œ ì €ì¥ë©ë‹ˆë‹¤.


````python
import os
import re
import json

# ğŸ“ 1ë‹¨ê³„: page_1.txtë¶€í„° page_12.txtê¹Œì§€ ë‚´ìš©ì„ í•˜ë‚˜ë¡œ í•©ì¹˜ê¸°
folder = "/content/output"
full_text = ""

for i in range(1, 13):
    file_path = os.path.join(folder, f"page_{i}.txt")
    if os.path.exists(file_path):
        with open(file_path, encoding="utf-8") as f:
            full_text += f.read() + "\n"  # âœ… ì¤„ë°”ê¿ˆê³¼ í˜•ì‹ì„ ê·¸ëŒ€ë¡œ ìœ ì§€í•¨
    else:
        print(f"âš ï¸ íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {file_path}")
````

````python
# "ìˆ«ì. ..." í˜•ì‹ìœ¼ë¡œ ì‹œì‘í•´ì„œ ë‹¤ìŒ ìˆ«ì ì•ê¹Œì§€ ë˜ëŠ” EOFê¹Œì§€ ë¸”ë¡ìœ¼ë¡œ ì¶”ì¶œ
pattern = r'(\d+\.\s.*?)(?=\n\d+\.\s|$)'
blocks = re.findall(pattern, full_text, flags=re.DOTALL)

all_questions = []  # ëª¨ë“  ë¬¸ì œë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
````

````python
for idx, block in enumerate(blocks, start=1):
    lines = block.strip().split("\n")

    # ì •ë‹µì´ ì‹œì‘ë˜ëŠ” ì¤„ ë²ˆí˜¸ ì°¾ê¸°
    ans_start = -1
    for i, line in enumerate(lines):
        if re.match(r'^(â‘ |â‘¡|â‘¢|â‘£|â‘¤|\d\s)', line.strip()):
            ans_start = i
            break

    if ans_start == -1 or ans_start < 1:
        print(f"âš ï¸ ë¬¸ì œ {idx}ì˜ ì •ë‹µ ë¶€ë¶„ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        continue

    # âœ… ë¬¸ì œ ë¶€ë¶„ ì •ë¦¬: ì¤„ë°”ê¿ˆì€ ìœ ì§€í•˜ê³ , ê¸€ë¨¸ë¦¬ ê¸°í˜¸ëŠ” ì œê±°
    ques_lines = lines[:ans_start]

    cleaned_lines = []
    buffer = ""

    for line in ques_lines:
        line = line.strip()

        if re.match(r'^[â€¢\-\â€“â—â–ªï¸â–¶ï¸â˜…â—†â– â˜‘ï¸âœ”ï¸â¤â”â–â£â¢â—‹]+\s*', line):
            # ì´ì „ ì¤„ì´ ì—°ê²° ì¤‘ì´ë¼ë©´ ì €ì¥
            if buffer:
                cleaned_lines.append(buffer.strip())
                buffer = ""
            # ê¸€ë¨¸ë¦¬ ê¸°í˜¸ ì œê±° í›„ ì¤„ ì €ì¥
            line = re.sub(r'^[â€¢\-\â€“â—â–ªï¸â–¶ï¸â˜…â—†â– â˜‘ï¸âœ”ï¸â¤â”â–â£â¢â—‹]+\s*', '', line)
            cleaned_lines.append(line.strip())
        else:
            # ì¼ë°˜ ì¤„ì´ë©´ ì´ì „ ì¤„ì— ì´ì–´ ë¶™ì´ê¸°
            buffer += " " + line

    if buffer:
        cleaned_lines.append(buffer.strip())

    question = "\n".join(cleaned_lines).strip()

    # ì •ë‹µ ì²˜ë¦¬
    ans_lines = lines[ans_start:]
    ans_raw = "\n".join(ans_lines)

    raw_ans = re.findall(
        r'(?:â‘ |1)\s?.*?(?=(?:â‘¡|2)\s|$)|'
        r'(?:â‘¡|2)\s?.*?(?=(?:â‘¢|3)\s|$)|'
        r'(?:â‘¢|3)\s?.*?(?=(?:â‘£|4)\s|$)|'
        r'(?:â‘£|4)\s?.*?(?=(?:â‘¤|5)\s|$)|'
        r'(?:â‘¤|5)\s?.*',
        ans_raw,
        flags=re.DOTALL
    )

    # ê° ë³´ê¸°ì—ì„œ ë²ˆí˜¸ ì œê±°
    answer = [
        re.sub(r'^(â‘ |â‘¡|â‘¢|â‘£|â‘¤|[1-5])[\s:.\-â€“]*', '', d.strip())
        for d in raw_ans
    ]

    all_questions.append({
        "question_id": idx,
        "question": question,
        "answer": answer
    })
````

````python
# ğŸ“ JSON íŒŒì¼ë¡œ ì €ì¥
with open("48th_exam_full.json", "w", encoding="utf-8") as f:
    json.dump(all_questions, f, ensure_ascii=False, indent=2)

print(f"\nâœ… ì™„ë£Œ: ì´ {len(all_questions)}ê°œì˜ ë¬¸í•­ì„ 12í˜ì´ì§€ì—ì„œ ì¶”ì¶œí–ˆìŠµë‹ˆë‹¤.")
````
